```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Creating list to append tweet data to
tweets_list2 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('Nation wide lockdown since:2020-01-01 near: "America"').get_items()):
    if i>10000:
        break
    tweets_list2.append([tweet.date, tweet.content, tweet.user.location])
    
# Creating a dataframe from the tweets list above
tweets_df2 = pd.DataFrame(tweets_list2, columns=['Datetime', 'Text'])
```

```python
tweets_df2
```
